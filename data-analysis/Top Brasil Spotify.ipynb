```python
import pandas as pd
import altair as alt
import numpy as np

data = pd.read_csv('../data/top_br.csv.zip')
data.head(10)
```

```python
data.fillna(0)

data.info(verbose=True)
```

```python
topartists = data['artist'].value_counts(ascending = False)[:10]
df = pd.DataFrame(topartists)
df.reset_index(level=0, inplace=True)
df.columns = ['artist','count']
df
```

```python
alt.Chart(df).mark_circle(color = '#222437').encode(
    x=alt.X('count', axis = alt.Axis(title='Popularidade Artista')),
    y=alt.Y('artist', axis = alt.Axis(title='Artists')),
    tooltip = ['count', 'artist']
)
```

```python
def gettop10track(column_name, show_name):
    data[column_name] = data[column_name].apply(pd.to_numeric, errors='coerce')
    a_df= data.drop_duplicates(['artist','name',column_name])[['artist','name',column_name]]
    #data[column_name] = data[column_name].astype('float')
    top10 = a_df.sort_values(column_name, ascending = False).head(10)
    
    chart = alt.Chart(top10).mark_circle(color = '#222437').encode(
    x = alt.X(column_name, axis=alt.Axis(title=show_name)),
    y = alt.Y('name:N',sort='-x', axis=alt.Axis(title='Música')),
    tooltip = [alt.Tooltip(column_name, title=show_name),
              alt.Tooltip('name:N', title='Música'),
              alt.Tooltip('artist:N', title='Artista')]
    )
    return chart                                                               
```

```python
gettop10track('danceability','Danceability')
```

```python
gettop10track('energy','Energy')
```

```python
gettop10track('acousticness','Acousticness')
```
